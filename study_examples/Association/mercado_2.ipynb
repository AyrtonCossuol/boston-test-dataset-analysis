# 2 - Importação da base

```
PATH = '/content/drive/MyDrive/Colab Notebooks/Machine Learning e Data Science com Python de A a Z/Estudo associação/mercado2.csv'
base = pd.read_csv(PATH, header = None)
base.head()
```

# 3 - Processamento e Resultados

## 3.1 - Transformando a base em matriz

```
transacoes = []
```

```
for i in range(0, base.shape[0]):
  transacoes.append([str(base.values[i, j]) for j in range(0, base.shape[1])])
```

## 3.2 - Utilizando o método apriori para fazer as devidas referência

```
regras = apriori(transacoes, 
                 min_support = 0.003, 
                 min_confidence = 0.2, 
                 min_lift = 3.0,
                 min_lenght = 2)
```

```
resultados = list(regras)
resultados
```

```
resultados_2 = [list(x) for x in resultados]
resultados_2
```

## 3.3 - Resultados encontrados

```
resultados_formatados = []
```

```
for j in range(0, 5):
  resultados_formatados.append([list(x) for x in resultados_2[j][2]])

resultados_formatados
```

 - **Conclusão**

 Utilizando um suporte mínimo de 0.003, junto com o lift de 3.0 e a confiança de 0.2, conseguimos obter os agrupamentos de produtos que são mais vendidos em conjunto.
